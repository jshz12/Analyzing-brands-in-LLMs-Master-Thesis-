```python
def randomization(mat):
  ### mat :  matrix where the results from the runs are stored, the numbers of the matrices are only 0 or 1 as we have dichotomized the LLMs responses.
  #Given a matrix and a chosen method of questioning, the columns are summed to derive the probability of responding 1 to that question.
  #Subsequently, the entropy of each question is calculated, and a vector of entropies is returned for each question format."


  epsilon = 1e-10
  n,m = mat.shape #n runs // m #questions

  #print("Runs:", n, "\nQuestions:", m)
  prob_pos = (np.sum(mat, axis=0))/n #total of 1 divided by number of runs.

  vec_entropy = []
  for i in range(m):
    p1 = prob_pos[i]
    p2 = 1-p1
    if p1 == 0:
      term1 = (p1 + epsilon) * np.log(p1 + epsilon)
    else:
      term1 = p1 * np.log(p1)

    if p2 == 0:
      term2 = (p2 + epsilon) * np.log(p2 + epsilon)
    else:
      term2 = p2 * np.log(p2)

    vec_entropy.append(-(term1+term2))#entropy per question

  return(vec_entropy)
```

## BIAS

```python
def sentiment_bias(mat):

  #Dada la matriz calculamos la suma por columnas y dividimos entre n. Probabilidad para 1 en la pregunta 1. Se repite para todas las preguntas y se saca la media
  #Tambien se devuelve la moda ( si hay mas 1 ó 0 )

  n,m = mat.shape #n runs // m #questions
  prob_pos = np.mean((np.sum(mat, axis=0))/n)
  #Para calcular la moda, sumamos todos los resultados de todas las preguntas. Si el valor es superior a la mitad de n*m entonces la moda es 1, otherwise 0
  tot_sum = sum(np.sum(mat,axis=0))
  if(tot_sum > n*m/2):
    mode = 1
  elif(tot_sum < n*m/2):
    mode = 0
  else:
    print("Can not decide whether the mode is 1 or 0. Same amount of each.")
    mode = -1

  return(prob_pos, mode,tot_sum)
```

## REFERENTIAL CONSISTENCY

```python
def kl_divergence(P,Q):
  epsilon = 1e-10
  for i in range(2):
      if P[i] == 0:
        P[i] = P[i] + epsilon
      if Q[i] == 0:
        Q[i] = Q[i] + epsilon
  kl_value = np.sum(P * np.log2(P / Q))  #logaritmo en base 2, asi conseguimos que la JS esté acotada
  return kl_value
```

```python
def jensen(mat1,mat2):

  n,m = mat1.shape #n runs // m #questions
  p = np.zeros(2)
  q = np.zeros(2)
  js_divergence = np.zeros(m)

  for j in range(m):
    p[0] = (np.sum(mat1[:,j], axis=0))/n #probability del 1
    p[1] = 1 - p[0] #probability del 0
    q[0] = (np.sum(mat2[:,j], axis=0))/n
    q[1] = 1 - q[0]
    #print(p[0],q[0])
    M = 0.5 * (p + q)
    js_divergence[j] = 0.5 *(kl_divergence(p, M)) + 0.5 *(kl_divergence(q, M))

  #print(js_divergence, "suma de toda la divergencia =", sum(js_divergence))
  return(js_divergence)
```

```python
def kullback(mat1,mat2):
  n,m = mat1.shape #n runs // m #questions
  p = np.zeros(2)
  q = np.zeros(2)
  kl_div = np.zeros(m)

  for j in range(m):
    p[0] = (np.sum(mat1[:,j], axis=0))/n #probability del 1
    p[1] = 1 - p[0] #probability del 0
    q[0] = (np.sum(mat2[:,j], axis=0))/n
    q[1] = 1 - q[0]
    #print(p[0],q[0])
    M = 0.5 * (p + q)
    kl_div[j] = kl_divergence(p,q)

  #print(kl_div, "suma de toda la divergencia =", sum(kl_div))
  return(kl_div)
```

```python
def entropia_por_pregunta(v1, v2, v3, v4, v5, v6, v7, v8,n, aux = 0):
    # Calcular la suma de los vectores para cada grupo
    entr_final_cs_1 = np.array(v1) + np.array(v2) + np.array(v3) + np.array(v4) #samsung
    entr_final_cs_2 = np.array(v5) + np.array(v6) + np.array(v7) + np.array(v8) #apple

    # Encontrar los índices de las n componentes con los valores más altos para cada grupo
    if aux == 0:
      top_indices_1 = np.argsort(entr_final_cs_1)[-n:]
      top_indices_2 = np.argsort(entr_final_cs_2)[-n:]
    else:
      top_indices_1 = np.argsort(entr_final_cs_1)[:n]
      top_indices_2 = np.argsort(entr_final_cs_2)[:n]


    # Coordenadas y para las barras
    y = np.arange(1, 101) #questions from 1 to 100
    bar_width = 0.5

    # Graficar las barras para cada componente de la pregunta
    for i in range(len(y)):
        if i in top_indices_1: #samsung
            plt.barh(y[i], -entr_final_cs_1[i], height=bar_width, color='red', alpha=0.7)
            plt.text(-entr_final_cs_1[i], y[i], f'{i+1}', ha='right', va='center', color='black', fontsize=6)
        else:
            plt.barh(y[i], -entr_final_cs_1[i], height=bar_width, color='blue', alpha=0.7)

        if i in top_indices_2:#apple
            plt.barh(y[i], entr_final_cs_2[i], height=bar_width, color='orange', alpha=0.7)
            plt.text(entr_final_cs_2[i], y[i], f'{i+1}', ha='left', va='center', color='black', fontsize=6)
          
        else:
            plt.barh(y[i], entr_final_cs_2[i], height=bar_width, color='green', alpha=0.7)

    # Fijar la posición del eje y en el centro
    plt.axvline(x=0, color='black', linewidth=1)

    plt.xlabel('Entropy')
    plt.ylabel('Question')
    plt.title('Samsung vs Apple')
    plt.tight_layout()
    plt.show()
```

```python
def plot_multiple_divergences(*divergences):
    """
    Función para graficar múltiples vectores de divergencias en un único cuadro.

    Args:
        *divergences: Uno o más vectores de divergencias.
    """
    x = range(1, 101)  # Eje x de 1 a 100

    plt.figure(figsize=(16, 9))  # Tamaño del cuadro de la figura
    vec_titles = ['ChatGPT3 Apple', 'ChatGPT3 Samsung', 'Gemma Apple', 'Gemma Samsung','ChatGPT4 Apple', 'ChatGPT4 Samsung']
    j= 0

    # Iterar sobre cada vector de divergencias y graficarlo
    for i, divergence in enumerate(divergences, start=1):
        plt.subplot(3, 2, i)  # Crear un subplot en una cuadrícula de 2x2
        # divergence = np.sort(divergence)
        plt.plot(x, divergence)

        # Añadir línea roja horizontal en el máximo valor de cada divergencia
        max_divergence = max(divergence)
        max_question = np.argmax(divergence)+1
        plt.axhline(y=max_divergence, color='r', linestyle='--', label=f'Maximum at question {max_question} ({max_divergence:.2f})')

        plt.xlabel('Question')
        plt.ylabel('Divergence')

        plt.title(f'{vec_titles[j]}')
        j = j+1
        plt.ylim(0, 0.5)
        plt.legend()  # Mostrar la leyenda con el valor máximo

    plt.tight_layout()
    plt.show()
```

```python
def draw_preference_triangle(pref_A_B, pref_A_C, pref_B_C, colordef ):
    # Coordenadas de los vértices del triángulo
    vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
    labels = ['Apple', 'Huawei', 'Samsung']
    
    # Configuración del gráfico
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.axis('off')  # Eliminar el cuadrado de los bordes

    # Dibujar el triángulo
    triangle = plt.Polygon(vertices, edgecolor='navy', linewidth=2, fill=True, facecolor='lightgray', alpha=0.7)
    ax.add_patch(triangle)
    
    # Añadir las marcas en los vértices con desplazamientos diferentes
    label_offsets = [-0.05, -0.05, +0.05]  # Desplazamientos para cada vértice
    
    for i, (x, y) in enumerate(vertices):
        ax.text(x, y + label_offsets[i], labels[i], fontsize=12, ha='center', va='center')
    
    # Función para añadir una flecha de preferencia
    def add_preference_arrow(start, end, preference, color):
        # Coordenadas inicial y final
        x_start, y_start = vertices[start]
        x_end, y_end = vertices[end]
        
        # Coordenadas de la flecha
        arrow_x = (1 - preference) * x_start + preference * x_end
        arrow_y = (1 - preference) * y_start + preference * y_end
        
        # Dibujar la línea de preferencia
        ax.plot([x_start, arrow_x], [y_start, arrow_y], color=color, lw=1.5)
        
        # Dibujar la línea de no preferencia (gris)
        ax.plot([arrow_x, x_end], [arrow_y, y_end], color='gray', lw=1.5)
        
        # Dibujar la flecha
        ax.annotate('', xy=(arrow_x, arrow_y), xytext=(x_start, y_start),
                    arrowprops=dict(arrowstyle="->", color=color, lw=3))

    # Añadir las flechas según las preferencias
    add_preference_arrow(1, 0, pref_A_B / 100, colordef)  # Preferencia de A sobre B
    add_preference_arrow(2, 0, pref_A_C / 100, colordef)  # Preferencia de A sobre C
    add_preference_arrow(1, 2, pref_B_C / 100, colordef)  # Preferencia de B sobre C
    
    # Mostrar el gráfico
    plt.show()
```

```python
# Define the function to plot
def entropy_function(x):
    # To avoid issues with log(0), we use np.where to handle x = 0 or x = 1
    return -(np.where(x == 0, 0, x * np.log(x)) + np.where(x == 1, 0, (1 - x) * np.log(1 - x)))

# Generate x values from 0 to 1
x_values = np.linspace(0, 1, 1000)

# Compute y values using the entropy function
y_values = entropy_function(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color = 'blue')
plt.axvline(x=0.5, color='r', linestyle='--')
plt.axvline(x=0.0, color='black')
plt.axhline(y=0, color='black')
plt.xlabel('p(x)')
plt.ylabel('Entropy')
plt.title('Entropy Function')
plt.legend()
plt.grid(True)

# Save the plot to a file
plt.savefig('entropy_plot.png')

# Show the plot
plt.show()
```
